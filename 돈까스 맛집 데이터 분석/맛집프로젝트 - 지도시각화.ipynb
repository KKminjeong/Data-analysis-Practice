```python
import pandas as pd
import googlemaps

gmaps = googlemaps.Client(key="고유 API 키 값") #각자 받으신 API키값을 넣으시면 됩니다.
```

```python
pork_cutlet_df = pd.read_excel('돈까스맛집리스트_210623_0951.xlsx',encoding = 'utf-8')
```

```python
#위도와 경도 값 추가
lat = []
lng = []

for name in pork_cutlet_df['주소']:
    tmpMap = gmaps.geocode(name)
    tmpLoc = tmpMap[0].get('geometry') 
    lat.append(tmpLoc['location']['lat']) 
    lng.append(tmpLoc['location']['lng'])
    
pork_cutlet_df['lat'] = lat
pork_cutlet_df['lng'] = lng

pork_cutlet_df.head(3)
```

```python
#별점 환산값 추가
def reRange(x, oldMin, oldMax, newMin, newMax):
    return (x - oldMin)*(newMax - newMin) / (oldMax - oldMin) + newMin 

pork_cutlet_df['망고환산'] = reRange(pork_cutlet_df['망고평점'], min(pork_cutlet_df['망고평점']), max(pork_cutlet_df['망고평점']), 1, 100)

pork_cutlet_df
```

```python
#다이닝코드별점이 없는 값을 포함시 전체 N/A값으로 되는 오류가 생겨 다이닝코드별점이 없는 값들을 제외시킨 후 환산

#다이닝코드별점 값만 가져오기


dining_df = pork_cutlet_df['다이닝코드별점']

dining_drop_df = dining_df.dropna()



# N/A값이 들어있는 행 삭제
# dining_drop_df = dining_df.dropna()

dining_drop_df

#환산 값 계산 후 df에 넣기


pork_cutlet_df['다이닝코드환산'] = reRange(dining_drop_df, min(dining_drop_df), max(dining_drop_df), 1, 100)

pork_cutlet_df
```

```python
#구별 식당수 구하기

#주소 없는 행 제거
pork_cutlet_df = pork_cutlet_df.dropna(subset=['주소'])

#서울시 자료만 가져오기
cutlet_seoul_df = pork_cutlet_df[pork_cutlet_df['주소'].str.startswith('서울')]

#구별 값 추출
cutlet_seoul_df['구별'] = cutlet_seoul_df['주소'].apply(lambda x : x.split(' ')[1])

#구별로 묶기(식당 수를 추가하기 위함)
cutlet_gu_df = pd.pivot_table(cutlet_seoul_df, index='구별')

#구별 음식점 수
cutlet_gu_df['식당수'] = cutlet_seoul_df['구별'].value_counts()

cutlet_gu_df
```

```python
import json

geo_path = 'skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
#구별 식당 개수 기준 시각화
import folium

map = folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='Stamen Toner')

map.choropleth(geo_data = geo_str, # 서울시 행정구역별 polygon drawing
               data = cutlet_gu_df['식당수'], # 시각화의 대상이 될 데이터
               columns = [cutlet_gu_df.index, cutlet_gu_df['식당수']], 
               fill_color = 'PuRd', 
               key_on = 'feature.id') 
map
```
