```python
import pandas as pd
#한글 형태소 분석위한 konlpy import
from konlpy.tag import Twitter

import re #정규 표현식 활용
```

```python
pork_cutlet_df = pd.read_excel('돈까스맛집리스트_210622_0924.xlsx',encoding = 'utf-8')
```

```python
pork_cutlet_df.head()
```

```python
def tokenizer(review):

    #이모티콘 제거 정규 표현식
    EMOJI = re.compile('[\U00010000-\U0010ffff]', flags=re.UNICODE)
    review_refine = EMOJI.sub(r'', review)
    
    #konlpy에서 형태소 분석 모듈 호출
    twitter = Twitter()
    
    #텍스트 형태소 분석
    raw_pos_tagged = twitter.pos(review_refine, norm = True, stem = True)
    
    #필요없는 단어 제거 리스트
    del_list = ['하다', '있다', '되다', '이다', '돼다', '않다', '그렇다', '아니다', '이렇다', '그렇다', '어떻다'] 

    word_cleaned = []

    #만든 del_list와 Konlpy에 있는 품사구분법 활용하여 명사, 형용사, 감탄사만 추출
    for word in raw_pos_tagged: #  ('서울', 'Noun'),
        if word[1] in ['Noun','Adjective','Exclamation']: # Foreign == ”, “ 와 같이 제외되어야할 항목들
            if (len(word[0]) != 1) & (word[0] not in del_list): # 한 글자로 이뤄진 단어들을 제외 & 원치 않는 단어들을 제외
                word_cleaned.append(word[0])

        
    word_dic = {}
    
    #list.count() 함수 활용하여 10번 이상 나타나는 단어들만 추출
    for word in word_cleaned:
        if word_cleaned.count(word) > 10:
            if word not in word_dic:
                word_dic[word] = 1 # changed from "0" to "1"
            else:
                word_dic[word] += 1

    return word_dic

```

```python
#토크나이징 루프
tokened_reviews=[]

for review in pork_cutlet_df['티스토리리뷰']:
    if review != '':
        #토크나이징 후 새로운 리스트에 추가
        t_review = tokenizer(review)
        tokened_reviews.append(t_review)


pork_cutlet_df['티스토리리뷰토큰'] = tokened_reviews

pork_cutlet_df.head()
```

```python
#하나의 문서에서 한 단어가 몇개 쓰였는지 f로 정의
def f(token, document):
    return document[token] #{'친구': 50, '경희대': 15, '돈카츠': 90, '맛집': 55, '인생...

#tf산출 공식
#t == token (하나의 토큰)
#d == document (하나의 문서 or dict) 
#D == Documents (전체문서 or list of dicts)

def tf(t, d):
    return 0.5 + 0.5*f(t, d)/max([ f(w, d) for w in d ])

tf('카츠',pork_cutlet_df['티스토리리뷰토큰'][0])
```

```python
#idf산출 공식
from math import log10

def idf(t, D):
    # D is documents == document list
    numerator = len(D)
    denominator = 1 + len([ True for d in D if t in d])
    return log10(numerator/denominator)

#tfidf 산출 공식
def tfidf(t, d, D):
    return tf(t,d)*idf(t, D)

#tf-idf 값 포함한 리스트 출력
def tfidfScorer(D):
    
    result = []
    for d in D:
        result.append([(t, tfidf(t, d, D)) for t in d])
    return result
```

```python
#tf-idf 공식 적용한 Top 5 키워드 점포별 선정

final_list = []

#tf-idf 값 포함한 리스트 
pork_cutlet_tfidf = tfidfScorer(pork_cutlet_df['티스토리리뷰토큰'])
# [(크림, 0.6937122209084516), (친구, 0.666666666666...]
# [(호점, 1.3010299956639813), (입구, 0.836376425783...]

#tf-idf 값을 key로 하여 Top 5만 내림차순 정렬한 리스트 생성
for token_list in pork_cutlet_tfidf:
    final_list.append(sorted(token_list, key=lambda x:x[1], reverse=True)[:5])

#tf-idf값 나타내는 새로운 Dataframe Column 생성
pork_cutlet_df['티스토리tfidf'] = final_list

top_five_list = []

#키워드만 추출하여 새로운 리스트 생성
for token_tuple_list in final_list:
    top_five = []
    for token_tuple in token_tuple_list:
        top_five.append(token_tuple[0])
    top_five_list.append(top_five)
        
#최종 신규 데이터프레임 열 생성 
pork_cutlet_df['탑5키워드(티스토리)'] = top_five_list
pork_cutlet_df.head(10)
```
