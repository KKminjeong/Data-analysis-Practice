```python
x_feature = 'Gr Liv Area'
y_feature = 'target'

X = df_train[x_feature]
Y = df_train[y_feature]

X
```

```python
# 각각의 평균 구하기
X_mean = np.mean(X)
Y_mean = np.mean(Y)

# 평균값을 이용하여 관계 구하기
relation = Y_mean/X_mean

print(f"{x_feature}의 평균:{X_mean:.2f} \n{y_feature}의 평균:{Y_mean:.2f}")
print(f"따라서 {y_feature}는 {x_feature}의 평균 {relation:.2f}배 입니다.")
print(f"즉, a = {relation:.2f} 입니다.")
a = round(relation, 2)
print(f"업데이트 된 방정식 : Y = {a:.2f} * X + b")
```

```python
# b = Y - a * X 를 그대로 작성. 평균값을 이용
b = Y_mean - a * X_mean

print(f"b는 {b:.2f} 입니다.")
print(f"업데이트 된 방정식 : Y = {a:.2f} * X + {b:.2f}")
```

```python
predict_Y = a * X + b
predict_Y
```

```python
def NMAE(true, pred):
    mae = np.mean(np.abs(true-pred))
    score = mae / np.mean(np.abs(true))
    return score
```

```python
# 실제 정답 값을 준비
real_answer = Y.copy()

# 정답과 예측 값을 함수에 넣어 결과를 확인
error = NMAE(real_answer, predict_Y)

print(f"이 방정식의 평균 에러는 {error:.2f} 입니다.")
```

```python
# test데이터 가져오기 
test_X = df_test[x_feature]

# 예측을 진행
predict_test = a * test_X + b
```
